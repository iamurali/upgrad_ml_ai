```python
import numpy as np
import collections
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

### Rewards

```python
with open('Rewards.pkl', 'rb') as handle:
    rewards_tracked = pickle.load(handle)   
```

```python
xaxis0 = np.asarray(range(0, len(rewards_tracked[(15,0)])))
xaxis1 = np.asarray(range(0, len(rewards_tracked[(25,1)])))
xaxis2 = np.asarray(range(0, len(rewards_tracked[(20,2)])))
xaxis3 = np.asarray(range(0, len(rewards_tracked[(30,3)])))
xaxis4 = np.asarray(range(0, len(rewards_tracked[(35,4)])))
xaxis5 = np.asarray(range(0, len(rewards_tracked[(10,5)])))
xaxis6 = np.asarray(range(0, len(rewards_tracked[(50,6)])))


plt.figure(0, figsize=(16,7))
plt.subplot(241)
plt.title('state=(15,mon)')
plt.plot(xaxis0,np.asarray(rewards_tracked[(15,0)]))
plt.subplot(242)
plt.title('state=(25,tue)')
plt.plot(xaxis1,np.asarray(rewards_tracked[(25,1)]))
plt.subplot(243)
plt.title('state=(20,wed)')
plt.plot(xaxis2,np.asarray(rewards_tracked[(20,2)]))
plt.subplot(244)
plt.title('state=(30,thu)')
plt.plot(xaxis3,np.asarray(rewards_tracked[(30,3)]))
plt.subplot(245)
plt.title('state=(35,fri)')
plt.plot(xaxis4,np.asarray(rewards_tracked[(35,4)]))
plt.subplot(246)
plt.title('state=(10,sat)')
plt.plot(xaxis5,np.asarray(rewards_tracked[(10,5)]))
plt.subplot(247)
plt.title('state=(50,sun)')
plt.plot(xaxis6,np.asarray(rewards_tracked[(50,6)]))
plt.show()
```
