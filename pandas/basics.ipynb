### Merge and append:

```python
gold = pd.DataFrame({'Country': ['USA', 'France', 'Russia'],
                         'Medals': [15, 13, 9]}
                    )
silver = pd.DataFrame({'Country': ['USA', 'Germany', 'Russia'],
                        'Medals': [29, 20, 16]}
                    )
bronze = pd.DataFrame({'Country': ['France', 'USA', 'UK'],
                        'Medals': [40, 28, 27]}
                    )
```

```python
pd.concat([gold, silver], axis=1)
```

```python
pd.concat([gold, silver], axis=0)
```

```python
gold.add(silver) # add rows parallely
```

```python
final_set = pd.concat([gold, silver, bronze])
result = final_set.groupby(['Country']).agg('sum').sort_values(['Medals'])
result
```

```python
gold.set_index('Country', inplace = True)
silver.set_index('Country', inplace = True) 
bronze.set_index('Country', inplace = True)
```

```python
gold.add(silver, fill_value=0).add(bronze, fill_value=0)
```

### _Lambda fn and pivot tables:_

```python
is_positive = lambda x: x>0
master_data['isProfit'] = master_data.Profit.apply(is_positive)
master_data.head()
```

```python
master_data.groupby('Customer_Segment').isProfit.mean()
```

```python
master_data['Profit'] = master_data['Profit'].apply(lambda x: round(x, 1))
master_data.head(2)
```

```python
master_data['profit_per_item'] = master_data['Profit'] / master_data['Order_Quantity']
master_data.head(2)
```
