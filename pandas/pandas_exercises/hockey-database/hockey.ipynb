```python
# set each team array with 0/1 based on the team assigned.

teams = set()
all_data['teams_array'] = all_data.Team.str.split("/")
for _teams in all_data['teams_array'].tolist():
    for _team in _teams:
        teams.add(_team)
        

for _team in teams:
    all_data[_team] = 0
    
        
def assign_values(x, t):
    if (t in x):
        return 1
    else:
        return 0


for _team in teams:
    all_data[_team] = all_data['teams_array'].apply(assign_values, args=([_team]))
```
