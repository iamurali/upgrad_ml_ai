### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
def gender_fn(x):
    if (x == 'M'):
        return 1
    else:
        return 0
    
users['gender_n'] = users['gender'].apply(gender_fn)

a = users.groupby('occupation').gender_n.sum() / users.occupation.value_counts() * 100
a.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').agg({'age': ['min', 'max']})
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
users_count = users.groupby(['occupation', 'gender']).gender.count()

occup_count = users.groupby(['occupation']).occupation.count()

users_count.div(occup_count, level= 'occupation') * 100
```
